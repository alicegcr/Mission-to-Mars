```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://redplanetscience.com/'
browser.visit(url)

html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
html_soup
```

```python
# html_soup.find('div', class_='content_title')
html_soup.select_one('div.content_title')
```

```python
html_soup.select('div.content_title')
```

```python
text = html_soup.find_all('div', class_='content_title')
text
```

```python
for i in text:
   # print(i.text)
    print(i.get_text())
```

```python
summary = html_soup.find_all('div', class_='article_teaser_body')
summary
```

```python
for div in summary:
    print(div.text)
```
